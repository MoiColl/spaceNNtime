# AADR divergence among samples

```python
import allel
import zarr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%load_ext rpy2.ipython
```

```python
%%bash

ls ../../data/AADR/
```

```python
ind = pd.read_table("../../data/AADR/v54.1_1240K_public_nospaces.ind", sep = "\t", names = ["ind", "sex", "pop"])
ind["idx"] = list(range(ind.shape[0]))
ind
```

```python
metadata = pd.read_table("../../files/AADR_filtered_metadata.txt", sep = "\t").query('sntind').query('sntind')
metadata["ind"] = metadata["indivi"]
metadata = metadata["ind"]
metadata
```

```python
ind_idx = pd.merge(ind, metadata, on='ind', how='inner').idx.to_numpy()
ind_idx
```

```python
snp = pd.read_table("../../data/AADR/v54.1_1240K_public_nospaces.snp", sep = "\t", names = ["id", "chrom", "genetic_pos", "pos", "ref", "alt"])
snp["idx"] = list(range(snp.shape[0]))
snp = snp.query("chrom == 1")
snp
```

```python
snp_idx = snp.idx.to_numpy()
snp_idx
```

```python
%%time

ac_gt = {0 : [2, 0],
         1 : [1, 1],
         2 : [0, 2],
         9 : [0, 0]}

geno = []
with open("../../data/AADR/v54.1_1240K_public.eigenstratgeno", "r") as file:
    for i, line in enumerate(file):
        if i in snp_idx:
            prev_geno = []
            for g in line.strip():
                prev_geno.append(int(g))
            prev_geno = np.array(prev_geno)[ind_idx]
            if np.sum(prev_geno == 9)/(prev_geno.shape[0]) < 0.25:
                geno.append(prev_geno)
        if i > 100000:
            break

geno = np.array(geno)
print(geno.shape)
geno
```

```python
geno_filt = np.copy(geno)
```

```python
homalt = np.sum(geno_filt == 2, axis = 1)
```

```python
hethet = np.sum(geno_filt == 1, axis = 1)
```

```python
homref = np.sum(geno_filt == 0, axis = 1)
```

```python
noncal = np.sum(geno_filt == 9, axis = 1)
```

```python
homalt+hethet+homref
```

```python
noncal
```

```python

for i in range(geno_filt.shape[0]):
    p = ((homalt[i]*2)+hethet[i])/((homalt[i]*2)+hethet[i]+(homref[i]*2))
    geno_filt[i, geno_filt[i, :] == 9] = np.random.binomial(n=2, p=p, size = noncal[i])
geno_filt
```

```python
def ld_prune(gn, size, step, threshold=.1, n_iter=1):
    for i in range(n_iter):
        loc_unlinked = allel.locate_unlinked(gn, size=size, step=step, threshold=threshold)
        n = np.count_nonzero(loc_unlinked)
        n_remove = gn.shape[0] - n
        print('iteration', i+1, 'snps', gn.shape[0], 'retaining', n, 'which is the ', round(n*100/gn.shape[0], 2),  '% removing', n_remove, 'variants')
        gn = gn.compress(loc_unlinked, axis=0)
    return gn
```

```python
gnu = ld_prune(geno_filt, size=500, step=200, threshold=.1, n_iter=4)
```

```python
coords, model = allel.pca(gnu, n_components=10, scaler='patterson')
```

```python
fig = plt.figure(figsize=(10, 5))

ax = fig.add_subplot(1, 2, 1)
ax.plot(coords[:, 0], coords[:, 1], marker='o', linestyle=' ', markersize=6, mec='k', mew=.5)
ax.set_xlabel('PC%s (%.1f%%)' % (0+1, model.explained_variance_ratio_[0]*100))
ax.set_ylabel('PC%s (%.1f%%)' % (1+1, model.explained_variance_ratio_[1]*100))
#ax.legend(bbox_to_anchor=(1, 1), loc='upper left')

ax = fig.add_subplot(1, 2, 2)
ax.plot(coords[:, 2], coords[:, 3], marker='o', linestyle=' ', markersize=6, mec='k', mew=.5)
ax.set_xlabel('PC%s (%.1f%%)' % (2+1, model.explained_variance_ratio_[2]*100))
ax.set_ylabel('PC%s (%.1f%%)' % (3+1, model.explained_variance_ratio_[3]*100))
#ax.legend(bbox_to_anchor=(1, 1), loc='upper left')

fig.suptitle("AADR PCA", y=1.02)
fig.tight_layout()
```

```python
fig = plt.figure(figsize=(10, 5))

ax = fig.add_subplot(1, 2, 1)
ax.plot(coords[:, 4], coords[:, 5], marker='o', linestyle=' ', markersize=6, mec='k', mew=.5)
ax.set_xlabel('PC%s (%.1f%%)' % (4+1, model.explained_variance_ratio_[4]*100))
ax.set_ylabel('PC%s (%.1f%%)' % (5+1, model.explained_variance_ratio_[5]*100))
#ax.legend(bbox_to_anchor=(1, 1), loc='upper left')

ax = fig.add_subplot(1, 2, 2)
ax.plot(coords[:, 6], coords[:, 7], marker='o', linestyle=' ', markersize=6, mec='k', mew=.5)
ax.set_xlabel('PC%s (%.1f%%)' % (6+1, model.explained_variance_ratio_[6]*100))
ax.set_ylabel('PC%s (%.1f%%)' % (7+1, model.explained_variance_ratio_[7]*100))
#ax.legend(bbox_to_anchor=(1, 1), loc='upper left')

fig.suptitle("AADR PCA", y=1.02)
fig.tight_layout()
```
